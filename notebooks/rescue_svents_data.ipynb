## Rescue svents data

Script to add sample_vents 'endtime', 're_intub_class' and 'time_on_vent' columns to patient chart events data if forgot include these columns in my sql query!

```python
import pandas as pd
```

```python
event = "tidalvolume"
data_dir = "../data/feathered/"
export_dir = "../data/feathered/"

datafile = data_dir+event
export_file = export_dir+event
```

```python
unfeathered_df = pd.read_feather(datafile)
```

```python
unfeathered_df.columns
```

```python
svents = pd.read_csv('../data/processed/sample_vents.csv')
```

```python
svents
```

```python
merged_inner = pd.merge(left=unfeathered_df, right=svents, left_on='hadm_id', right_on='hadm_id')
```

```python
merged_inner.drop(['index'],axis=1,inplace=True)
```

```python
merged_inner.rename(columns={'subject_id_x':'subject_id'},inplace=True)
```

```python
merged_inner
```

```python
to_save = merged_inner[['subject_id', 'hadm_id', 'stay_id', 'charttime', 'storetime',
       'itemid', 'value', 'valuenum', 'valueuom', 'warning','endtime','re_intub_class','time_on_vent']]
```

```python
to_save
```

```python
to_save.to_feather(export_file)
```
