## Getting age and sex data

#### import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### get feature dataset

```python
df = pd.read_csv('all_feature_table.csv')
```

#### load patient dataset containing age and sex

```python
patients = pd.read_csv('patients.csv')
```

#### get only patients who have been ventilalted from patient data set

```python
ids = df['subject_id']
vented_patients = patients[patients['subject_id'].isin(ids)]
```

```python
vented_patients
```

```python
dup_hadms = df['subject_id'].value_counts()
```

```python
dup_hadms
```

```python
df[df['subject_id'].isin(dup_hadms.index[dup_hadms >10])]
```

#### join age and gender to main feature table and save to .csv

```python
new_df = pd.merge(left = df, right=vented_patients, how = 'left', left_on='subject_id', right_on='subject_id') 
```
